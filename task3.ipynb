¿Qué riesgo corre si establece un umbral demasiado estricto (ej. 0.5px)? ¿Qué pasa con la matriz final si el umbral es muy laxo (ej. 50px)?

**R//** *Si el umbral es demasiado estricto, el riesgo es que se rechacen muchos puntos que en realidad sí son correctos, pero tienen pequeños errores por ruido o imprecisiones en la detección, esto hace que termine siendo inestable o poco confiable.*

*Pero si el umbral es muy laxo, muchos puntos incorrectos se aceptan como si fueran buenos. Entonces la matriz final se calcula usando datos erróneos y se deforma, produciendo una mala alineación o distorsiones en la imagen final.*

2. Imagine que usted es el Lead Computer Vision Engineer de una empresa de drones. Deben alinear imágenes térmicas de paneles solares tomadas desde el aire para detectar fallos.

a. Problema: El drone vuela a 50 metros de altura. El terreno no es perfectamente plano (hay colinas suaves), pero los paneles sí son planos.

**R//** *No es válido usar una homografía global porque el terreno no es completamente plano. Como hay colinas, cambia la profundidad y una sola matriz no puede alinear bien toda la imagen. Solo funcionaría bien si todo fuera plano, como los paneles.*

b. Pregunta A: ¿Es válido usar una Homografía global para unir todo el mapa del terreno? ¿Por
qué sí o por qué no?

**R//** *Para reducir el tiempo, primero eliminaría la mayor cantidad posible de matches incorrectos antes de aplicar RANSAC. Así habría menos outliers y el algoritmo necesitaría menos iteraciones. También se puede detener antes cuando ya se encuentra un buen modelo.*

c. Pregunta B: Su algoritmo RANSAC está tardando demasiado (3 segundos por frame) en la computadora a bordo del drone (Raspberry Pi). La telemetría indica que el 90% de los matches iniciales son outliers debido al pasto y árboles repetitivos.


i. Proponga una estrategia concreta para reducir el tiempo de ejecución sin cambiar el hardware. (Pista: Piense en la fórmula de N o en pre-filtrado geométrico)

**R//** *Como el 90% de los puntos son incorrectos, RANSAC necesita muchas pruebas para encontrar puntos buenos y por eso tarda mucho. Una solución podría ser eliminar primero la mayor cantidad posible de puntos malos antes de usar RANSAC. Así habrá más puntos buenos y el algoritmo necesitará menos iteraciones, reduciendo el tiempo.*
